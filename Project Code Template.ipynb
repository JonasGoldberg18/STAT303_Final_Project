```python
# Merge all divvy data into one large dataset - Rebecca Lazarus
divvy_data = pd.concat([divvy_data_2020, divvy_data_2021, divvy_data_2022, divvy_data_2023, divvy_data_2024])
```

```python
# Check for missing values - Rebecca Lazarus
divvy_data.isnull().sum()
```

```python
#Impute start_station_name missing values - Rebecca Lazarus
divvy_data['lat_round'] = divvy_data['start_lat'].round(3)
divvy_data['lng_round'] = divvy_data['start_lng'].round(3)
station_name_reference = divvy_data.dropna(subset=['start_station_name'])
sation_name_lookup = dict(zip(zip(station_name_reference['lat_round'], station_name_reference['lng_round']), station_name_reference['start_station_name']))

def fill_station(row):
    if pd.notnull(row['start_station_name']):
        return row['start_station_name']
    else:
        return sation_name_lookup.get((row['lat_round'], row['lng_round']))

divvy_data['start_station_name'] = divvy_data.apply(fill_station, axis=1)

divvy_data.isnull().sum()
```

```python
#Check type of remaining missing values - Rebecca Lazarus
missing_data = divvy_data[divvy_data['start_station_name'].isnull()]
print(missing_data['rideable_type'].value_counts())
```

```python
#Impute remaining start_station_name values - Rebecca Lazarus
divvy_data.loc[
    (divvy_data['start_station_name'].isnull()) & (divvy_data['rideable_type'] == 'electric_bike'), 
    'start_station_name'
] = 'Public Rack'

divvy_data['start_station_name'] = divvy_data['start_station_name'].fillna('Unknown')
divvy_data.drop(columns=['lat_round', 'lng_round'], inplace = True)

divvy_data.isnull().sum()
```

```python
#Impute end_station_name missing values - Rebecca Lazarus
divvy_data['lat_round'] = divvy_data['end_lat'].round(3)
divvy_data['lng_round'] = divvy_data['end_lng'].round(3)
end_station_name_reference = divvy_data.dropna(subset=['end_station_name'])
end_sation_name_lookup = dict(zip(zip(end_station_name_reference['lat_round'], end_station_name_reference['lng_round']), end_station_name_reference['end_station_name']))

def fill_end_station(row):
    if pd.notnull(row['end_station_name']):
        return row['end_station_name']
    else:
        return sation_name_lookup.get((row['lat_round'], row['lng_round']))

divvy_data['end_station_name'] = divvy_data.apply(fill_end_station, axis=1)

divvy_data.isnull().sum()
```

```python
#Check type of remaining missing values - Rebecca Lazarus
missing_data = divvy_data[divvy_data['end_station_name'].isnull()]
print(missing_data['rideable_type'].value_counts())
```
