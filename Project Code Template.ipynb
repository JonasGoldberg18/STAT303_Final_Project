```python
# traffic is congestion is either high/med/low
# this is determined based on the speed
# if speed is in bottom 33%, then there is high
# congestion (slow speed). If speed is in 66%, then it is 
# medium congestion (medium speed). Otherwise there is low 
# congestion (high speed traffic flow)

# returns .33 and .66 percentiles 
# for speed column
def percentile33(x):
    return np.quantile(x, 1/3) #1/3
def percentile66(x):
    return np.quantile(x, 2/3)
```

```python
# get percentiles for each region: 
speed_per_region = traffic.groupby('REGION_ID')['SPEED'].agg([percentile33, percentile66])
speed_per_region = speed_per_region.rename(columns={"percentile33": 'low_speed', "percentile66": "medium_speed"})
speed_per_region
```

```python
# we will determine if a bike route is in a traffic station 
# based on the start location of the bike. 
# To determine if a bike station is in a traffic region: 
# WEST <= start_lng <= EAST and
# SOUTH <= start_lat <= NORTH

# first let's get unique regions 
duplicates = traffic['REGION_ID'].duplicated()
duplicates = ~duplicates
regions = traffic[duplicates]
```

```python
# OPTIMIZED REGION FINDER: 
bools = []

for i in range(29): # 29 if going up
    west = regions.iloc[i].loc['WEST']
    east = regions.iloc[i].loc['EAST']
    south = regions.iloc[i].loc['SOUTH']
    north = regions.iloc[i].loc['NORTH']
    lng_indices = (west <= divvy_data['start_lng']) & (divvy_data['start_lng'] <= east)
    lat_indices = (south <= divvy_data['start_lat']) & (divvy_data['start_lat'] <= north)
    indices = lng_indices & lat_indices
    bools.append(indices)

# multiple = np.array(bools)
final_index = pd.Series(np.array([-1] * len(bools[0])))

# final_index = bools[0]


for id, index in enumerate(bools):
    indices = final_index < 0
    final_index[index.to_numpy() & indices.to_numpy()] = id + 1

indices = final_index < 0
final_index[indices] = pd.NA
```

```python
divvy_data['region_id'] = final_index

# Filter NANs
nulls = divvy_data['region_id'].isnull()
divvy_data = divvy_data[~nulls]

# next, we need to find the traffic congestion time that is 
# closest to ridership 'started_at' time. This is to get 
# the traffic congestion measurement and then extract the 
# speed (to determine if there is high/low congestion)
```

```python
divvy_data['started_at'] = pd.to_datetime(divvy_data['started_at'], errors='coerce')
traffic['TIME'] = pd.to_datetime(traffic['TIME'])

# filter null dates
null_dates = divvy_data['started_at'].isnull()
divvy_data = divvy_data[~null_dates]
```

```python
# sort the dates. Sorting will allow us to efficiently execute an algorithim that will be explained later
divvy_data = divvy_data.sort_values("started_at").reset_index(drop=True)
traffic = traffic.sort_values("TIME").reset_index(drop=True)
```

```python
# we need to find the traffic time stamp that is closest to each divvy time stamp. 
# This is because we want to get the traffic estimate at the closest time to when
# someone begins their bike ride. 

# Due to the sheer number of rows in both datasets, iterating through each one and finding the minimum 
# is too costly. As a result, sorting the time stamps makes it more efficient because the closest 
# time stamp will be equal to or greater than the index of the previous closest time stamp. 

# NAIVE METHOD (takes too long)
# mins = []
# traffic_times = traffic['TIME'].values
# divvy_times = divvy_data['started_at'].values

# for i in range(divvy_data.shape[0]):
#     diff = np.abs(traffic_times - divvy_times[i])
#     mins.append(diff.argmin())


# OPTIMAL APPROACH: 
def find_min(traffic_times, divvy_point, curr_min):
    min_diff = abs(traffic_times[curr_min] - divvy_point)
    for i in range(curr_min, traffic_times.shape[0]):
        curr_diff = abs(traffic_times[i] - divvy_point)
        if curr_diff < min_diff:
            min_diff = curr_diff
            curr_min = i
        if curr_diff > min_diff:
            return curr_min
    return curr_min

divvy_times = divvy_data['started_at'].astype(int).to_numpy()
traffic_times = traffic['TIME'].astype(int).to_numpy()

# first find the current minimum naively: 
curr_min = np.abs(divvy_times[0] - traffic_times).argmin()
mins = [curr_min]
```

```python
for i in range(1, divvy_times.shape[0]):
    curr_min = find_min(traffic_times, divvy_times[i], curr_min)
    mins.append(curr_min)


mins = np.asarray(mins)
divvy_data['speed'] = traffic['SPEED'][mins].reset_index()['SPEED']

```

```python
# finally, we determine based on the speed if there is 
# low/medium/high congestion
def congestion_rate(row):
    id = row['region_id']
    if row['speed'] <= speed_per_region.loc[id, 'low_speed']:
        return 'high'
    if row['speed'] <= speed_per_region.loc[id, 'medium_speed']:
        return 'medium'
    return 'low'

congestion_vals = divvy_data.apply(congestion_rate, axis=1)
divvy_data['congestion'] = congestion_vals
```

```python
#expressing Race as Percentages

totals = race_data_clean.iloc[0]
race_data_clean = race_data_clean / totals
race_data_clean = race_data_clean * 100

race_data_clean = race_data_clean.drop("Total:")
```

```python
#findind the latitude and longitude of every Zip code in the census data that we use
lat_long_data = pd.read_excel("../Census_Data/long_lat_zip.xlsx", header=0)


lat_long_data['zip'] = 'ZCTA5 ' + lat_long_data['zip'].astype(str)
lat_long_data = lat_long_data[lat_long_data['zip'].isin(census_data.index)].reset_index(drop=True)
lat_long_data = lat_long_data.loc[:, ['zip', 'lat', 'lng']]
lat_long_data.head()
```

```python
#finding the closest census zip code for each start lat and long in the 2020 divvy data (I will be using euclidean distance for this)

divvy_lats = divvy_2020['start_lat'].to_numpy().astype(float)
divvy_longs = divvy_2020['start_lng'].to_numpy().astype(float)

zip_lats = lat_long_data['lat'].to_numpy().astype(float)
zip_longs = lat_long_data['lng'].to_numpy().astype(float)

divvy_lats  = divvy_lats.reshape(-1, 1)
divvy_longs = divvy_longs.reshape(-1, 1)

zip_lats = zip_lats.reshape(1, -1)
zip_longs = zip_longs.reshape(1, -1)

dist_sq = (divvy_lats - zip_lats)**2 + (divvy_longs - zip_longs)**2
nearest_idx = dist_sq.argmin(axis=1)

zip_codes = lat_long_data['zip'].to_numpy()
divvy_2020['start_zip'] = zip_codes[nearest_idx]

divvy_2020.head()
```

```python
#Counting How many rides started in each ZCTA
zip_counts = divvy_2020.groupby('start_zip').size().reset_index(name='ride_count')
zip_counts = zip_counts.set_index('start_zip')
zip_counts.head()

census_data['divvy_ride_count'] = zip_counts['ride_count']
census_data.head()
```
